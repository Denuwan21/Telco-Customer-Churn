```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
Telco = pd.read_csv("Telco-Customer-Churn.csv")
Telco
```

```python
Telco.isnull()
```

```python
Yes_No_Maping_1 = {"Yes" : 1, "No" : 0}
Yes_No_Maping_2 = {"Yes" : 1, "No" : 0, "No internet service" : 0}
Yes_No_Maping_3 = {"Yes" : 1, "No" : 0, "No phone service" : 0}
Yes_No_Maping_4 = {"DSL" : 1, "Fiber optic" : 2, "No" : 0}
Yes_No_Maping_5 = {"Month-to-month" : "MM", "One year" : "OY", "Two year" : "TY"}
Yes_No_Maping_6 = {"Electronic check" : "E-Check","Bank transfer (automatic)" : "Banck", "Mailed check" : "M-Check", "Credit card (automatic)" : "Card"}

Telco["Partner"] = Telco["Partner"].map(Yes_No_Maping_1)
Telco["Dependents"] = Telco["Dependents"].map(Yes_No_Maping_1)
Telco["PhoneService"] = Telco["PhoneService"].map(Yes_No_Maping_1)
Telco["PaperlessBilling"] = Telco["PaperlessBilling"].map(Yes_No_Maping_1)
Telco["Churn"] = Telco["Churn"].map(Yes_No_Maping_1)
Telco["OnlineSecurity"] = Telco["OnlineSecurity"].map(Yes_No_Maping_2)
Telco["OnlineBackup"] = Telco["OnlineBackup"].map(Yes_No_Maping_2)
Telco["DeviceProtection"] = Telco["DeviceProtection"].map(Yes_No_Maping_2)
Telco["TechSupport"] = Telco["TechSupport"].map(Yes_No_Maping_2)
Telco["StreamingTV"] = Telco["StreamingTV"].map(Yes_No_Maping_2)
Telco["StreamingMovies"] = Telco["StreamingMovies"].map(Yes_No_Maping_2)
Telco["MultipleLines"] = Telco["MultipleLines"].map(Yes_No_Maping_3)
Telco["InternetService"] = Telco["InternetService"].map(Yes_No_Maping_4)
Telco["Contract"] = Telco["Contract"].map(Yes_No_Maping_5)
Telco["PaymentMethod"] = Telco["PaymentMethod"].map(Yes_No_Maping_6)

Sex = pd.get_dummies(Telco["gender"], drop_first = True)
D_Contract = pd.get_dummies(Telco["Contract"], drop_first = True)
Payment = pd.get_dummies(Telco["PaymentMethod"], drop_first = True)

Telco = pd.concat([Telco, Sex, D_Contract , Payment], axis = 1)

Telco.drop(["customerID", "gender", "PaymentMethod", "Contract", "TotalCharges"], axis = 1, inplace = True)

Telco
```

```python
X = Telco.drop(["Churn"], axis = 1)
Y = Telco.Churn
```

```python
Telco["Male"] = Telco["Male"].astype(int)
Telco["OY"] = Telco["OY"].astype(int)
Telco["TY"] = Telco["TY"].astype(int)
Telco["Card"] = Telco["Card"].astype(int)
Telco["E-Check"] = Telco["E-Check"].astype(int)
Telco["M-Check"] = Telco["M-Check"].astype(int)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
Result = LogisticRegression()
Result.fit(X_train, Y_train)
```

```python
Result.score(X_train, Y_train)
```
